## Source Text (to summarize)

Prompt engineering is the practice of designing instructions that guide an AI system to produce useful, reliable outputs.
Even small changes in wording can affect accuracy, tone, and whether the model follows constraints. A strong prompt
usually specifies the role the assistant should play, the task goal, and the format of the desired output. It can also
include constraints, such as maximum length, required sections, or what information to avoid requesting.

One common technique is prompt chaining. Instead of asking for everything at once, you break a task into multiple steps
where the output of one step becomes the input to the next. For example, a customer support agent might first classify a
ticket (billing vs. login vs. bug), then ask only for missing details, then propose a solution, and finally decide whether
to escalate to a human agent. Chaining improves reliability because each step is smaller and easier to verify. It also
creates a clear record of how the assistant reached its final response.

Another technique is ReACT prompting, which blends reasoning with actions. A ReACT workflow often looks like: plan the
approach, write code, run it, observe results or errors, then fix and rerun. This is especially useful in coding tasks
because execution reveals real issues (like typos, missing imports, or incorrect assumptions). By making each stage explicit,
ReACT helps the model and the user debug systematically instead of guessing.

Self-reflection prompts (also called critique-and-revise) ask the AI to evaluate its own output against explicit criteria,
then produce an improved version. Effective criteria include accuracy (does it match the source), clarity (is it easy to read),
audience fit (is it appropriate for the reader), and constraints (word limit, bullet format, required points). The key is
to force a revision, not just a critique, so the final output improves in measurable ways.

Across these techniques, iteration matters. You rarely get the best result in one attempt. Testing prompts on realistic
inputs, noting failures, and refining constraints is part of the process. When documenting work for grading, it helps to
show the prompts used, the outputs produced, and at least one improvement cycle (before/after) so a reviewer can see that
you tested and refined the approach rather than writing prompts only once.


## Step B — Self-Critique Prompt (explicit criteria)

Copy/paste this prompt as your **self-reflection / critique request** (this is what you’re graded on).

**Revised Summary constraints**
- Accuracy (must reflect the source)
- Clarity (simple language)
- Audience: high school senior / non-expert
- Length limit: **max 120 words**
- Format: **bullet points**
- Must mention: **prompt chaining, ReACT, self-reflection, and iteration**
- Must include: **exactly 2 actionable tips** (imperative sentences)
- Output must include two sections with headings: `Critique` and `Revised Summary`


```python

self_critique_prompt = """
You are reviewing a summary for a non-expert reader.

TASK:
1) Critique the summary using these criteria:
- Accuracy (matches the source)
- Clarity (easy to read)
- Audience fit (high school senior)
- Constraints (word limit, required concepts)
2) Then produce a revised version that meets ALL constraints:

CONSTRAINTS FOR REVISED SUMMARY:
- Max 120 words
- Bullet points only
- Must mention: prompt chaining, ReACT, self-reflection, iteration
- Must include exactly 2 actionable tips (imperative sentences)
- Output format must be:

Critique:
- ...
- ...
Revised Summary:
- ...
- ...
"""

print(self_critique_prompt)
```


## Step C — Improved Summary (AFTER)

Run the cell below to show an example critique + revised summary that meets the constraints.


```python

critique = [
    "Accuracy: Mostly correct, but it doesn’t clearly separate the three techniques or explain why each helps.",
    "Clarity: A bit dense; using bullets and simpler phrasing would improve readability.",
    "Constraints: No word limit enforcement, no bullet format, and no explicit actionable tips."
]

revised_summary = [
    "Prompt engineering is writing instructions that shape an AI’s output (role, goal, and format).",
    "Prompt chaining breaks one task into steps so each output feeds the next and is easier to verify.",
    "ReACT is a loop (plan → code → run → observe → fix) that helps debug by using execution results.",
    "Self-reflection critiques output against criteria and then revises it for measurable improvement.",
    "Iteration matters because testing and refining prompts increases reliability over time.",
    "Use prompt chaining to split complex tasks into checkable steps.",
    "Test prompts on realistic inputs and tighten constraints when outputs drift."
]

print("Critique:")
for b in critique:
    print("-", b)

print("\nRevised Summary:")
for b in revised_summary:
    print("-", b)

word_count = sum(len(b.split()) for b in revised_summary)
print(f"\nApprox word count (revised bullets only): {word_count}")
```
