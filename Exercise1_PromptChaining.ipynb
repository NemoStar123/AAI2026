# Exercise 1 — Prompt Chaining for Customer Support AI

**Tool Used:** Google Colab  
**Scenario:** TaskFlow the SaaS productivity app

Customer message:
> "I was charged twice and now I can't log in. Please help."

Prompt Chain:
1. Classify issue  
2. Ask missing info  
3. Propose solution  
4. Escalation rule  

Each step uses the previous step's output.



## Step 1 — Classification Prompt

Classify the issue into:
billing / login / bug / other

Return JSON:
{
  "category": "",
  "confidence": 0.0,
  "tags": []
}


```python

# Example Step 1 Output
step1_output = {
    "category": "billing",
    "confidence": 0.95,
    "tags": ["double-charge", "login-failure"]
}
step1_output
```


## Step 2 — Gather Missing Information

Ask up to 5 necessary questions based on Step 1 output.
Do not request sensitive data.
Return numbered list.


```python

step2_questions = [
    "What email is associated with your account?",
    "What date and amount were the duplicate charges?",
    "What exact error appears when logging in?",
    "Have you tried resetting your password?",
    "Are you using a specific browser or device?"
]

step2_questions
```


## Step 3 — Proposed Solution

Draft a response:
- Empathetic tone
- Max 200 words
- Step-by-step actions
- Mention refund timeline (5–7 business days if confirmed)


```python
step3_solution = """
Hi — I’m sorry you’re experiencing this. Let’s fix both issues.

1) Try logging in using an incognito/private window.
2) Clear your browser cache and retry.
3) Reset your password using “Forgot Password.”

Regarding billing: once we confirm the duplicate charge, we will issue a refund within 5–7 business days.

If this does not resolve your issue, reply "Escalate" and we will prioritize your case.
"""

print(step3_solution)
```


## Step 4 — Escalation Rule

Return JSON:
{
  "escalate": true/false,
  "priority": "",
  "route_to": "",
  "conditions_triggered": []
}


```python

step4_escalation = {
    "escalate": True,
    "priority": "High",
    "route_to": "Billing + Tier 2 Support",
    "conditions_triggered": [
        "Duplicate charge reported",
        "Login issue persists after reset"
    ]
}

step4_escalation
```
