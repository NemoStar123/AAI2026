# Exercise 2 — Code Generation with ReACT Prompting

Tools Used: Google Colab, GBT

## Goal
Use a ReACT-style cycle:
Reason → Act → Run → Observe → Fix → Re-run

Task:
1. Loaded a CSV of pre made student grades
2. Compute per-student average
3. Flag students with average < 80 as "At Risk"
4. Demonstrate an error and fix it



## ReACT Prompt Structure

**Reason / Plan**
- Load student_grades.csv
- Compute average across math, science, english
- Handle missing values
- Flag students below 80

**Act**
Generate Python code.

**Run**
Execute the code.

**Observe**
Capture spelling errors.

**Fix**
Correct the issue and rerun.


```python

# Create student_grades.csv

import pandas as pd
import numpy as np

data = {
    "student": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "math": [85, 78, 92, 88, 76],
    "science": [90, 74, 89, 85, np.nan],
    "english": [88, 82, 94, 79, 75]
}

df = pd.DataFrame(data)
df.to_csv("student_grades.csv", index=False)
df
```


## Run Buggy Code (Intentional Error)
This demonstrates iteration evidence.


```python

import pandas as pd

df_bug = pd.read_csv("student_grades.csv")

# Intentional typo: 'sciene'
df_bug["average"] = df_bug[["math", "sciene", "english"]].mean(axis=1)

df_bug["status"] = df_bug["average"].apply(lambda x: "At Risk" if x < 80 else "OK")

df_bug
```


## Observation
The previous cell raises a KeyError due to a misspelled column name.
We will fix it below.


```python

df_fixed = pd.read_csv("student_grades.csv")

df_fixed["average"] = df_fixed[["math", "science", "english"]].mean(axis=1, skipna=True)

df_fixed["status"] = df_fixed["average"].apply(lambda x: "At Risk" if x < 80 else "OK")

df_fixed
```


## Status Counts


```python

df_fixed["status"].value_counts()
```


## Notes

Before: Column name was misspelled ('sciene'), causing KeyError.

After: Corrected to 'science' and used skipna=True to handle missing values.

 Demonstrates the full ReACT cycle and successful execution.
